##1 The mean sales generated by each zone

```python
Zone=[]

for i in dataset.columns[1:5]:
  Zone.append([i,dataset[i].mean()])
Z_M=pd.DataFrame(Zone,columns=['Zone','Mean-Sales'])
Z_M
```

```python
plt.figure(figsize=(10,7))
plt.bar(Z_M['Zone'],Z_M['Mean-Sales'],width=.8,color='indigo')
plt.xlabel('Zone')
plt.ylabel('Mean-Sales')
plt.title("Mean Sales Zone Wise")
plt.plot()
```

#3 Check whether all the zones generate the same amount of sales.

####H0-null hypothesis:All zones generate same sales
####H1-alternate hypothesis:All zones do not generate same sales

```python
f,p=st.f_oneway(df['Zone - A'],df['Zone - B'],df['Zone - C'],df['Zone - D'])
```

```python
f,p
```

we see that the p value is less than .05, so we will reject the null hypothesis and accept the alternate hypothesis. 

we can conclude that all the zones generate different sales
